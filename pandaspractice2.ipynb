# Importing movies dataset

```python
import pandas as pd
df = pd.read_csv("movies.csv", low_memory=False)
df
```

```python
df.info()
```

# Dropping irrelevant columns

```python
df.drop(columns= ['adult', 'video', 'homepage', 'poster_path', 'status', 'production_companies', 'production_countries', 'belongs_to_collection'], inplace = True)
df
```

# Converting string representations of lists to actual lists

```python
import ast
import numpy as np

df['genres'] = df['genres'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else np.nan)
df['spoken_languages'] = df['spoken_languages'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else np.nan)

df
```

# Converting dict format to easily readable string format

```python
df['genres'] = df['genres'].apply(lambda x: ", ".join(i['name'] for i in x) if isinstance(x, list) else x)
df['genres'].replace("", np.nan)

df['spoken_languages'] = df['spoken_languages'].apply(lambda x: ", ".join(i['name'] for i in x) if isinstance(x, list) else x)
df['spoken_languages'].replace("", np.nan)

df
```

```python
df.isna().sum()
```

# Converting and cleaning select columns 

```python
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
df['budget'] = df['budget'].replace(0, np.nan)

df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')
df['revenue'] = df['revenue'].replace(0, np.nan)

df.rename(columns={'budget' : 'budget_USD', 'revenue' : 'revenue_USD'}, inplace = True)

df
```

```python
df['id'] = pd.to_numeric(df['id'], errors = 'coerce')
df['popularity'] = pd.to_numeric(df['popularity'], errors = 'coerce')
```

# Rearranging order of columns (movie title shifted to first column)

```python
column_name = 'original_title'
df = df[[column_name] + [col for col in df.columns if col != column_name]]

df
```

```python
df['release_date'] = pd.to_datetime(df['release_date'], errors = 'coerce')
```

# Dealing with missing data using replace function

```python
df['overview']
df['overview'].replace('No overview', np.nan, inplace = True)
df['overview'].replace('No overview yet', np.nan, inplace = True)
df['overview'].replace('No overview available', np.nan, inplace = True)
df['overview'].replace(" ", np.nan, inplace = True)
```

# Removing duplicates

```python
df = df.drop_duplicates()
df = df.reset_index(drop=True)
```

# Creating plots with lines of best fit. Analyzing correlation between budget & popularity, and budget & revenue

```python
import matplotlib.pyplot as plt
import seaborn as sns

df_clean = df.dropna(subset=['budget_USD', 'popularity', 'revenue_USD'])

corr_budget_popularity = df_clean['budget_USD'].corr(df_clean['popularity'])
print(f"Correlation between Budget and Popularity: {corr_budget_popularity}")

plt.figure(figsize=(8, 6))
sns.regplot(x='budget_USD', y='popularity', data=df_clean, scatter_kws={'s': 10, 'alpha': 0.1}, line_kws={'color': 'red'})
plt.title('Budget vs Popularity with Line of Best Fit')
plt.xlabel('Budget (USD)')
plt.ylabel('Popularity')
plt.grid(True)
plt.show()
```

```python
corr_budget_revenue = df_clean['budget_USD'].corr(df_clean['revenue_USD'])
print(f"Correlation between Budget and Revenue: {corr_budget_revenue}")

plt.figure(figsize=(8, 6))
sns.regplot(x='budget_USD', y='revenue_USD', data=df_clean, scatter_kws={'s': 10, 'alpha': 0.1}, line_kws={'color': 'green'})
plt.title('Budget vs Revenue with Line of Best Fit')
plt.xlabel('Budget (USD)')
plt.ylabel('Revenue (USD)')
plt.grid(True)
plt.show()
```
